# plotting the graph for "Temperature Variation Over Time"

```python
# Converting 'Date/Time' column to datetime format
data['Date/Time'] = pd.to_datetime(data['Date/Time'])

# Plotting the graph
plt.figure(figsize=(10, 6))
plt.plot(data['Date/Time'], data['Temp_C'], color='blue', marker='o', linestyle='-')
plt.title('Temperature Variation Over Time')
plt.xlabel('Date/Time')
plt.ylabel('Temperature (Â°C)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

# plotting the graph to find the relation between temperature and wind speed

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Temp_C'], data['Wind Speed_km/h'], color='green', alpha=0.5)
plt.title('Temperature vs. Wind Speed')
plt.xlabel('Temperature (Â°C)')
plt.ylabel('Wind Speed (km/h)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# plotting the graph to observe the occurance of clear weather over time 

```python
# Createing a binary column indicating whether the weather is clear or not
data['Clear_Weather'] = data['Weather'].apply(lambda x: 1 if x == 'Clear' else 0)

# Plotting the graph
plt.figure(figsize=(10, 6))
plt.plot(data['Date/Time'], data['Clear_Weather'], color='blue', marker='o', linestyle='-')
plt.title('Occurrence of "Weather is Clear" Over Time')
plt.xlabel('Date/Time')
plt.ylabel('Clear Weather (1 = Clear, 0 = Not Clear)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
